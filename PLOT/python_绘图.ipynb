```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.mlab as mlab
import matplotlib.font_manager as fm
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 解决中文乱码
matplotlib.rcParams['font.sans-serif'] = ['SimHei']  # 乱码
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y_sin, y_cos = np.sin(x), np.cos(x)
    
plt.figure(figsize=(8, 6), dpi=80)
plt.title('plot1')
plt.grid()
plt.xlabel('x轴', fontdict={'fontsize': 15})
plt.ylabel('y轴', fontdict={'fontsize': 15})

plt.xlim(-4, 4)
plt.ylim(-1, 1)

plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.yticks(np.linspace(-1, 1, 9, endpoint=True))

plt.plot(x, y_cos, 'b--', lw=2.0, label='cos(x)')
plt.plot(x, y_sin, 'g--', lw=2.0, label='sin(x)')

plt.legend(loc='best', shadow=True)

plt.show()
```

1. 将坐标轴移向中心

```python
ax = plt.gca()
# 1. 将右边和上边的轴的颜色设置为无
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# 2. 分别将左边的和下边的设置为y轴， x轴
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# 坐标原点（0， 0）
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c, s = np.cos(x), np.sin(x)

plt.plot(x, c, c='r', lw=2.0, ls='--', label='cos(x)')
plt.plot(x, s, c='g', lw=2.0, ls='-', label='sin(x)')

plt.legend(loc='best')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'0', r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 1])

plt.show()
```

条形图

```python
means_men = [20, 35, 30, 35, 27]
means_women = [25, 32, 34, 20, 25]

title('柱状图')

index = arange(len(means_men))

bar(index, means_men, width=.3, alpha=.6, color='r', label='男生')
bar(index+.3, means_women, width=.3, alpha=.6, color='g', label='女生')
legend(loc='best', shadow=True)
plot(index, means_men, 'r*-')
plot(index+.3, means_women, 'g+-')

for i in range(len(means_men)):
    text(index[i], means_men[i]+0.3, means_men[i], ha='center', va='bottom')
    text(index[i]+.3, means_women[i]+0.3, means_women[i], ha='center', va='bottom')
xticks([x + 0.15 for x in index], ['A', 'B', 'C', 'D', 'E'])
show()
```

横向柱状图

```python
barh(index, means_men, height=.3, color='g', label='Men')
barh(index+.3, means_women, height=.3, color='b', label='women')

for i in range(len(means_men)):
    text(means_men[i]+.3, index[i], means_men[i], va='center', ha='left')
    text(means_women[i]+.3, index[i]+.3, means_women[i], va='center', ha='left')
yticks([x+.15 for x in index], ['A', 'B', 'C', 'D', 'E'])
show()
```

层次柱状图

```python
data = np.array([
    [1, 4, 2, 5, 2],
    [2, 1, 1, 3, 6],
    [5, 3, 1, 4, 6]
])

title('层次柱状图')
index = arange(len(data[0]))
colors = ['r', 'g', 'b']

bottom = array([0, 0, 0, 0, 0])

for i in range(len(data)):
    bar(index, data[i], width=.5, color=colors[i], bottom=bottom, alpha=.7, label=''+str(i))
    bottom += data[i]
legend(loc='best')
show()
    
```

直方图

```python
mu, sigma = 100, 5
x = mu + sigma*np.random.randn(10000)
title('直方图')
n, bins, patches = hist(x, bins=50, normed=1, cumulative=False, color='g', alpha=.8, label='直方图')

y = mlab.normpdf(bins, mu, sigma)
plot(bins, y, 'r--')

legend(loc='best')

show()
```

饼状图

```python
sizes = [15, 30, 45, 10]
labels = ["Frogs", "中文", "Dogs", "Logs"]
colors = ["yellowgreen", "gold", "lightskyblue", "lightcoral"]
explode = [0, 0.05, 0, 0]
title('饼状图')

patches, l_text, p_text = pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
axis('equal')

show()
```

填充图

```python
x = linspace(-2*pi, 2*pi, 1000, endpoint=True)
y = sin(x)

title('填充图')
plot(x, y, c='b', alpha=.7)
# 填充图形, plt.fill_between(x, y1, y2, where=None, *kwargs)
fill_between(x, 0, y, where=(y>0), color='r', alpha=.3)
fill_between(x, 0, y, where=(y<0), color='g', alpha=.3)

show()
```

雷达图

```python
labels = np.array(["A组", "B组", "C组", "D组", "E组", "F组"])
data = np.array([68, 83, 90, 77, 89, 73])
theta = np.linspace(0, 2*np.pi, len(data), endpoint=False)

# 构成一个封闭图形
data = concatenate((data, [data[0]]))  # array([68, 83, 90, 77, 89, 73, 68])
theta = concatenate((theta, [theta[0]]))   

subplot(111, polar=True)
title('雷达图')

thetagrids(theta*(180/pi), labels=labels)
rgrids(arange(60, 100, 20), labels=arange(60, 100, 20), angle=0)   # 间隔

plot(theta, data, 'bo--', lw=2)
fill(theta, data, color='r', alpha=.4)

show()
```
