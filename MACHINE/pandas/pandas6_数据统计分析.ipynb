<h3>数据切割</h3>

```python
import pandas as pd 
import numpy as np
```

```python
pd.cut(range(10), bins=[0, 2, 4, 6, 8, 10], right=False)
```

```python
data = np.random.randint(0, 1000, size=10000)
```

```python
data_cut= pd.cut(data, bins=[0, 200, 400, 600, 800, 1000], right=False)
data_cut = data_cut.count_values()
```

```python
print(data_cut)
```
