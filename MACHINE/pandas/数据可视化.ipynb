### 调整subplot周围的间距

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=.5)
# wspace和hspace用于控制宽度和高度的百分比，可以用作subplot之间的间距
plt.subplots_adjust(wspace=0, hspace=0)
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='default')
plt.plot(data, 'k--', drawstyle='steps-post', label='step-post')
plt.legend(loc='best')
```

### 在图表中添加一个图形

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

### 使用pandas和seaborn绘图

```python
import pandas as pd
```

```python
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'],
                 index=np.arange(0, 100, 10))
df.plot(figsize=(12, 9))
```

```python
plt.figure(figsize=(12, 9))
fig, ax = plt.subplots(2, 1)
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(ax=ax[0], color='g', alpha=.7, figsize=(12, 9))
data.plot.barh(ax=ax[1], color='r', alpha=.4, figsize=(12, 9))
```

```python
# plt.figure(figsize=(12, 9))
df = pd.DataFrame(np.random.rand(6, 4),
                 index=['1', '2', '3', '4', '5', '6'],
                 columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df.plot.bar(figsize=(12, 9))
```

```python
df.plot.barh(stacked=True, alpha=.6, figsize=(12, 9))
```

```python
data = pd.Series(np.random.randn(1000))
data.plot.hist(bins=50, figsize=(12, 9), density=1)
data.plot.density(style='g--')
# plt.show()
```
