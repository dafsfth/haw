```python
data = pd.read_csv('土木补打卡名单.xlsx', chunksize=20)
tot = pd.Series([])
for d in data:
    tot = tot.add(d['key'].value_counts(), fil_value=0)
tot = tot.sort_values(ascending=False)
tot[:10]
```

### json数据

```python
obj = """
{"name": "Wes",
     "places_lived": ["United States", "Spain", "Germany"],
     "pet": null,
     "siblings": [{"name": "Scott", "age": 30, "pets": ["Zeus", "Zuko"]},
                  {"name": "Katie", "age": 38,
                   "pets": ["Sixes", "Stache", "Cisco"]}]}
"""
```

通过json.loads即可将JSON字符串转换成Python形式：

```python
import json
result = json.loads(obj)
result
```

json.dumps则将Python对象转换成JSON格式：

```python
obj_json = json.dumps(result)
```

```python
data2 = pd.DataFrame(data=result['siblings'], columns=['name', 'age', 'pets'])
```

### 数据库

```python
import sqlite3
```

```python
query = """
       CREATE TABLE test(a VARCHAR(20), b VARCHAR(20),c REAL, d INTEGER);
       """
```

```python
con = sqlite3.connect('mydata.sqlite3')
```

```python
con.execute(query)
```

```python
con.commit()
```

```python
data = [('Atlanta', 'Georgia', 1.25, 6),
        ('Tallahassee', 'Florida', 2.6, 3),
        ('Sacramento', 'California', 1.7, 5)]
stmt = 'INSERT INTO test VALUES(?, ?, ?, ?)'
con.executemany(stmt, data)
```

```python
cursor = con.execute('select * from test')
rows = cursor.fetchall()
rows
```

```python
pd.DataFrame(rows, columns=[x[0] for x in cursor.description])
```
